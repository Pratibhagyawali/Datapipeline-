# New York Housing Price Analysis

## 1. Introduction
This project analyzes housing data from New York.  
The goal is to understand how price per square foot changes across different locations in the city.  
We clean the dataset, calculate new values, and create visualizations to explore patterns.

## 2. Importing Libraries
We load the Python libraries needed for data analysis and visualization.
- pandas for working with tables  
- numpy for numerical operations  
- matplotlib for charts  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

## 3. Loading the Dataset
We load the CSV file into a pandas DataFrame and show the first few rows to confirm it loaded correctly.

```python
df = pd.read_csv("NY-House-Dataset.csv")
df.head()
```

## 4. Cleaning the Data
We convert price and square footage to numeric values, remove invalid rows, and create a new column for price per square foot.


```python
# Convert columns to numeric
df["PRICE"] = pd.to_numeric(df["PRICE"], errors="coerce")
df["PROPERTYSQFT"] = pd.to_numeric(df["PROPERTYSQFT"], errors="coerce")

# Remove rows with missing or invalid values
df = df.dropna(subset=["PRICE", "PROPERTYSQFT"])
df = df[df["PROPERTYSQFT"] > 0]

# Create price per square foot
df["price_per_sqft"] = df["PRICE"] / df["PROPERTYSQFT"]

df.head()
```

```python
df["price_per_sqft"].describe()
```

## 5. Scatter Plot: Longitude vs Price per Sqft
This plot shows how price per square foot changes from east to west across New York.


```python
plt.figure(figsize=(8,5))
plt.scatter(df["LONGITUDE"], df["price_per_sqft"], s=5, c="red", alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Price per Sqft")
plt.title("Longitude vs Price per Sqft")
plt.show()
```

## 6. Scatter Plot: Latitude vs Price per Sqft
This plot shows how price per square foot changes from south to north.


```python
plt.figure(figsize=(8,5))
plt.scatter(df["LATITUDE"], df["price_per_sqft"], s=5, c="blue", alpha=0.5)
plt.xlabel("Latitude")
plt.ylabel("Price per Sqft")
plt.title("Latitude vs Price per Sqft")
plt.show()
```

## 7. Creating Price Bins
We divide price per square foot into five groups (bins) from cheapest (0) to most expensive (4).


```python
df["price_bin"] = pd.qcut(df["price_per_sqft"], 5, labels=[0,1,2,3,4])
df["price_bin"].value_counts().sort_index()
```

## 8. Map-Style Plot
This plot shows New York like a map, with colors representing different price levels.


```python
plt.figure(figsize=(6,6))
scatter = plt.scatter(
    df["LONGITUDE"],
    df["LATITUDE"],
    c=df["price_bin"].astype(int),
    cmap="viridis",
    s=5,
    alpha=0.6
)
plt.colorbar(scatter, label="Price bin (0 = cheapest)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("New York Apartments by Price per Sqft (Binned)")
plt.show()
```

## 9. Conclusion
We observe that higher price per square foot is concentrated in specific areas, while cheaper properties are found further from the central high-demand zones. Location clearly has a strong impact on housing prices in New York.
